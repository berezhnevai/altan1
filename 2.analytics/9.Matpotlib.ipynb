- Диаграммы рассеивания

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.scatter(x, y)

plt.show()
```

```python
x = np.random.randn(500)
y = np.random.randn(500)

# colors = np.random.random(500)
# colors = np.linspace(0, 1, 500)
colors = x

# подбор размера точек

sizes = np.random.random(500) * 100
plt.scatter(x, y, c=colors, cmap= "plasma", s=sizes)

plt.show()
```

```python
# данные по населению
p = {
    "Yakutsk":318768, 
    "Neryungri": 57009,
    "Viliusk": 11095
}

a = {
    "Yakutsk":122, 
    "Neryungri": 98.9,
    "Viliusk": 15
}

# координаты (широта и долгота)
lat = {
    "Yakutsk":62, 
    "Neryungri": 56,
    "Viliusk": 63
}
lon = {
    "Yakutsk":129, 
    "Neryungri": 124,
    "Viliusk": 121
}
data = pd.DataFrame(
    {"population":p, "area": a, "latitude": lat, "longtitude": lon}
)
data
```

```python
# визуализация данных
# логарифмирование данных по населению 
pop_log = np.log10(data["population"])

plt.scatter(data["longtitude"],data["latitude"], s=data["area"], c = pop_log)

# цыетовая шкала

plt.colorbar(label="население (логарифм)")

plt.show()
```

```python
x = np.random.randn(500)
y = np.random.randn(500)

# colors = np.random.random(500)
# colors = np.linspace(0, 1, 500)
colors = x

# подбор размера точек

sizes = np.random.random(500) * 100
symbol = "irii"
plt.scatter(x, y, c=colors, cmap= "plasma", s=sizes, marker = f"${symbol}$")

plt.show()
```

```python
# объединение в одном окне графика и диаграммы рассеивания

x = np.linspace(0, 10, 1000)
y = np.sin(x)

# модуляция
z = y + np.random.randn(1000) * 0.5

# диаграмма рассеивания
plt.scatter(x, z, c = np.linspace(z.min(), z.max(), 1000), cmap = 'seismic')

# график
plt.plot(x, y)

plt.show()
```
