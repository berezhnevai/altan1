# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

простые линейные графики

```python
# 1-мерный массив (вектор) из 10 значений (аналог математического вектора) 
x = np.arange(10)
x
```

```python
y = x
```

```python
# визуализация
plt.plot(x, y)
plt.show()
```

```python
y = x ** 2

plt.plot(x, y)
plt.show()
```

```python
# 1 -мерный массив из 1000 значений
# линейная последовательность в диапазоне от 0 до 10 
x = np.linspace(0, 10, 1000)
x
```

```python
# данные по оси x
x = np.linspace(-10, 10, 1000)

# определение фунцкии
def sigmoid(arg):
    return 1 / (1+np.exp(-arg))

# вычисление значений по оси y
y = sigmoid(x)
# визуализация
plt.plot(x, y)
plt.show()
```

```python
# гиперболический тангес
y = np.tanh(x)
plt.plot(x, y)
plt.show()
```

```python
# x = np.linspace(-10,10,10)
x = np.linspace(-10,10,1000)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
# совмещение некскольких графиков в одном окне
x = np.linspace(-10,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = sigmoid(x)
y4 = np.tanh(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.show()
```

```python
# цвет линий 
plt.plot(x, y1, color="red") # по названию
plt.plot(x, y2, color="m") # по первой букве цвета из форматов rgb 
plt.plot(x, y3, color="#ffa70360", linewidth = 10) #по 16-ричному коду формата RGB
plt.plot(x, y4, color=(0.0, 1.0, 0.5)) #по формату RGB в виде кортежа со значениями в диапазоне от 0 до 1

plt.show()
```

```python
# стиль линий
plt.plot(x, np.sin(x), linestyle='-', color="red") #"solid") #сплошная линия
plt.plot(x, np.sin(x - 1), linestyle='--', color="black") #"dashed") #штриховая линия
plt.plot(x, np.sin(x - 2), linestyle=':') #"") #пунктирная линия
plt.plot(x, np.sin(x - 3), linestyle='-.') #"") #штрих-пунктирнаz
plt.show()
```

```python
# объединение настроек цвета и стиля линий

plt.plot(x, np.sin(x), "--g")
plt.plot(x, np.sin(x+1), ":r")
plt.plot(x, np.sin(x-1), "-.y")
plt.show()
```

```python
# пределы осей
plt.plot(x, np.sin(x), "--g")

# индивидуальные пределы

plt.xlim(-15, 15)
plt.ylim(-1.5, 1.5)

plt.axis([-20,20,-1.5,1.5])
plt.show()
```

```python
# Легенда графика


plt.plot(x, np.sin(x), "--g", label="синус")
plt.plot(x, np.cos(x), ":r", label="косинус")
plt.legend()

plt.axis([-20,20,-2,2])
plt.show()
```

```python
# Подписи оси


plt.plot(x, np.sin(x), "--g", label="синус")
plt.plot(x, np.cos(x), ":r", label="косинус")
plt.legend()

plt.xlabel("x")
plt.ylabel("sin and cos")


plt.axis([-20,20,-2,2])
plt.show()
```

```python
# Заголовок графика


plt.plot(x, np.sin(x), "--g", label="синус")
plt.plot(x, np.cos(x), ":r", label="косинус")
plt.legend()

plt.xlabel("x")
plt.ylabel("sin and cos")
plt.title("Графики синуса и косинуса")

plt.axis([-20,20,-2,2])
plt.show()
```
